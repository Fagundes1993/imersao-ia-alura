Configurar Nome do Bucket e Banco de Dados:
Crie um bucket no Google Cloud Storage para armazenar as imagens.
Defina o nome do bucket e o nome do arquivo do banco de dados:

```python
bucket_name = "simulado"  # Substitua pelo nome do seu bucket
db_name = "questoes.db"
```

Funções para Upload de Imagem e Criação do Banco de Dados:
Crie as funções upload_imagem() e criar_banco_dados():

```python
def upload_imagem(arquivo_local, nome_arquivo):
    """Faz upload de uma imagem para o Cloud Storage e retorna o URL público."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(nome_arquivo)
    blob.upload_from_filename(arquivo_local)
    return blob.public_url

def criar_banco_dados():
    """Cria o banco de dados SQLite e a tabela de questões."""
    conn = sqlite3.connect(db_name)
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS questoes
                  (assunto TEXT,
                   imagem_url TEXT,
                   gabarito TEXT)''')
    conn.commit()
    conn.close()
```

Função para Cadastrar Questões:
Crie a função cadastrar_questao():

```python
def cadastrar_questao(assunto, imagem_local, gabarito):
    """Cadastra uma questão no banco de dados."""
    imagem_url = upload_imagem(imagem_local, f"{assunto}_{gabarito}.jpg")
    conn = sqlite3.connect(db_name)
    c = conn.cursor()
    c.execute("INSERT INTO questoes VALUES (?, ?, ?)", (assunto, imagem_url, gabarito))
    conn.commit()
    conn.close()
```

Criar o Banco de Dados:
Execute a função para criar o banco de dados (se ele ainda não existir):

```python
criar_banco_dados()
```

Cadastrar Questões:
Use a função cadastrar_questao() para adicionar questões ao banco de dados. Substitua os valores de exemplo pelos seus dados:

```python
assunto = "Políticas Públicas"
imagem_local = "/content/questoes/pb01.png"  # Caminho para a imagem que você enviou
gabarito = "C"
cadastrar_questao(assunto, imagem_local, gabarito)
```
